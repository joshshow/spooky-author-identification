```python
import tensorflow as tf
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, Dense, LSTM
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

print(train.shape)
print(test.shape)
```

```python
train.head()
```

```python
print(train.value_counts("author"))
```

```python
toki = Tokenizer(num_words=10000)
toki.fit_on_texts(train["text"])
le = LabelEncoder()
```

```python
X = toki.texts_to_sequences(train["text"])
X = pad_sequences(X, maxlen = 100)
y = le.fit_transform(train['author'])

print(X.shape, y.shape)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.layers import Dropout

ls_fn = SparseCategoricalCrossentropy(from_logits=True)

es = EarlyStopping(
    monitor='val_loss',
    patience = 2,
    restore_best_weights=True
)

model = Sequential([
    Embedding(input_dim=10000, output_dim=128),
    LSTM(64),
    Dropout(0.5),
    Dense(units=3, activation='linear')
])    

model.compile(loss=ls_fn, optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10, callbacks=[es], validation_split=0.2)
```

```python
nony = model(X)
y_pred = tf.nn.softmax(nony).numpy()
```

```python
test.head()
```

```python
X_test = toki.texts_to_sequences(test['text'])
X_test = pad_sequences(X_test, maxlen=100)

nonyy = model(X_test)
answer = tf.nn.softmax(nonyy).numpy()
```

```python
submissions = pd.DataFrame(answer, columns=["EAP", "HPL", "MWS"])

submissions['id'] = test['id']
submissions = submissions[['id', 'EAP', 'HPL', 'MWS']]
```

```python
submissions.to_csv("mySubmission.csv", index=False)
```
